```python
import pandas as pd
from Bio.Seq import Seq
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading in whitelisted results and calculating ratio + stdev

```python
whitelist_results_with_step1 = pd.read_csv("../../output/TL4_umi_results_whitelist_concat_custom_threshold/results_with_step1.csv")
whitelist_results_with_step1["AD_prot"] = whitelist_results_with_step1["AD"].map(lambda s: str(Seq(s).translate()))
whitelist_results_with_step1
```

```python
whitelist_results_with_step1["complex_ratio"] = whitelist_results_with_step1["RTBC_umi_count_complex"] / whitelist_results_with_step1["AD_umi_count_complex"]
whitelist_results_with_step1 = whitelist_results_with_step1[
    np.isfinite(whitelist_results_with_step1["complex_ratio"])
]
whitelist_results_with_step1
```

```python
whitelist_results_with_step1_agg = whitelist_results_with_step1[["AD", "complex_ratio", "name_RPTR"]].groupby(["AD",  "name_RPTR"]).agg([np.mean, np.std]).reset_index()
whitelist_results_with_step1_agg.columns = ["AD", "name_RPTR", "mean", "std"]
whitelist_results_with_step1_agg
```

```python
sns.set_context('talk')
ax = sns.scatterplot(
    data=whitelist_results_with_step1_agg,
    x="mean",
    y="std",
    edgecolor='none', alpha = 0.5,s =10
)

# ax.set_xscale('log')
# ax.set_yscale('log')

# add y = x line
lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),
    np.max([ax.get_xlim(), ax.get_ylim()])
]
ax.plot(lims, lims, '--', color='gray')
ax.set_xlim(lims)
ax.set_ylim(lims)
ax.set_xlabel("Mean Ratio")
ax.set_ylabel("Activity Standard Deviation")
sns.despine()
```

# Compare to before whitelisting

```python
results_with_step1_no_ec = pd.read_csv("../../output/TL4_umi_results/results_with_step1.csv", index_col = 0)
results_with_step1_no_ec["complex_ratio"] = results_with_step1_no_ec["RTBC_umi_count_complex"] / results_with_step1_no_ec["AD_umi_count_complex"]
results_with_step1_no_ec["AD_prot"] = results_with_step1_no_ec["Designed"].map(lambda s: str(Seq(s).translate()))
results_with_step1_no_ec = results_with_step1_no_ec[
    np.isfinite(results_with_step1_no_ec["complex_ratio"])
]

results_with_step1_no_ec_agg = results_with_step1_no_ec[["Designed", "complex_ratio", "name_RPTR"]].groupby(["Designed",  "name_RPTR"]).agg([np.mean, np.std]).reset_index()
results_with_step1_no_ec_agg.columns = ["AD", "name_RPTR", "mean", "std"]
results_with_step1_no_ec_agg
```
