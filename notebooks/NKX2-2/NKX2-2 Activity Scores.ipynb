```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Puro only

```python
AD_BC_results = pd.read_csv("../../output/NKX2-2/TL4B2/puro_only/no_err_corr/AD_trebl_experiment_results.csv", index_col = 0)
AD_BC_results["time"] = AD_BC_results["name"].str.split("_").str[5].str[1:].astype(int)
AD_BC_results["rep"] = AD_BC_results["name"].str.split("_").str[4].str[1:].astype(int)
AD_BC_results  = AD_BC_results[["ADBC2", "HawkBCs", "time","rep", "AD_umi_count_simple"]].reset_index(drop = True)
AD_BC_results
```

```python
RT_BC_results = pd.read_csv("../../output/NKX2-2/TL4B2/puro_only/no_err_corr/RT_trebl_experiment_results.csv")
RT_BC_results["time"] = RT_BC_results["name"].str.split("_").str[5].str[1:].astype(int)
RT_BC_results["rep"] = RT_BC_results["name"].str.split("_").str[4].str[1:].astype(int)
RT_BC_results  = RT_BC_results[["RTBC", "time", "rep","RTBC_umi_count_simple"]].reset_index(drop = True)
RT_BC_results
```

```python
step1_map = pd.read_csv("../../output/NKX2-2/pipeline_test/no_err_corr/step1.csv")
step1_map = step1_map[["ADBC2", "HawkBCs", "RTBC", "AD"]]
step1_map
```

```python
step1_map_with_AD = pd.merge(step1_map, AD_BC_results)
step1_map_with_AD
```

```python
step1_map_with_RT = pd.merge(step1_map, RT_BC_results)
step1_map_with_RT
```

```python
step1_map_with_AD_RT = pd.merge(step1_map_with_AD, step1_map_with_RT, how = 'outer')
step1_map_with_AD_RT["AD_umi_count_simple"] = step1_map_with_AD_RT["AD_umi_count_simple"].fillna(0)
step1_map_with_AD_RT["RTBC_umi_count_simple"] = step1_map_with_AD_RT["RTBC_umi_count_simple"].fillna(0)
step1_map_with_AD_RT
```

```python
step1_map_with_AD_RT["activity"] = step1_map_with_AD_RT["RTBC_umi_count_simple"] / step1_map_with_AD_RT["AD_umi_count_simple"]
step1_map_with_AD_RT
```

```python
# step1_map_with_AD_RT = step1_map_with_AD_RT.drop(columns = ["AD_umi_count_simple", "RTBC_umi_count_simple"])
# step1_map_with_AD_RT
```

```python
grouped = step1_map_with_AD_RT.groupby(["AD", "time", "rep"])["activity"].agg(["mean", "std"]).reset_index()
grouped
```

```python
pivoted_grouped_activity = grouped.pivot(index=["AD", "rep"], columns="time", values="mean")
pivoted_grouped_activity
```

```python
pivoted_grouped_activity.to_csv("../../output/NKX2-2/TL4B2/puro_only/no_err_corr/trebl_experiment_activities.csv")
```

```python
pivoted_grouped_activity['na_count'] = pivoted_grouped_activity.isna().sum(axis=1)
```

```python
pivoted_grouped_sd = grouped.pivot(index=["AD", "rep"], columns="time", values="std")
pivoted_grouped_sd['na_count'] = pivoted_grouped_sd.isna().sum(axis=1)
pivoted_grouped_sd
```
