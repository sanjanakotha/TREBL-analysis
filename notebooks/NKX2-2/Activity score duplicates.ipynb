```python
dup_rows = puro_no_err_corr_trebl_exp_bc[
    puro_no_err_corr_trebl_exp_bc
        .groupby(["rep", "time", "AD", "RTBC_umi_count_simple"])["AD"]
        .transform("size") > 1
]

dup_rows = dup_rows.sort_values(by = ["AD", "rep", "time"])
dup_rows
```

```python
problem_cases_with_bcs = (
    puro_no_err_corr_trebl_exp_bc
        .groupby(["rep", "time", "AD", "RTBC"])
        .agg(
            n_rows=("AD", "size"),
            n_ADBC2=("ADBC2", "nunique"),
            n_HawkBC=("HawkBCs", "nunique"),
            ADBC2s=("ADBC2", lambda x: sorted(x.unique())),
            HawkBCs=("HawkBCs", lambda x: sorted(x.unique())),
        )
        .reset_index()
        .query("n_ADBC2 > 1 or n_HawkBC > 1")
)

problem_cases_with_bcs
```
