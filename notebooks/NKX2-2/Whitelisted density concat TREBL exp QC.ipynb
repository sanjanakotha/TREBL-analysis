```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Before whitelisting barcodes?

```python
results_with_step1 = pd.read_csv("../../output/TL4_umi_results/results_with_step1.csv", index_col = 0)
results_with_step1 
```

```python
results_with_step1["complex_ratio"] = results_with_step1["RTBC_umi_count_complex"] / results_with_step1["AD_umi_count_complex"]
results_with_step1
```

```python
results_with_step1_agg = results_with_step1[["Designed", "complex_ratio", "name_RPTR"]].groupby(["Designed",  "name_RPTR"]).agg([np.mean, np.std]).reset_index()
results_with_step1_agg.columns = ["AD", "name_RPTR", "mean", "std"]
results_with_step1_agg
```

```python
# Most ADs have reasonable error
import seaborn as sns
sns.set_context('talk')
plt.figure(dpi = 300)
ax = sns.scatterplot(data = results_with_step1_agg, 
                     x = "mean", 
                     y = "std", 
                     edgecolor = 'none', alpha = 0.5)
sns.despine()

# Add y = x line
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
lims = [min(xmin, ymin), max(xmax, ymax)]
ax.plot(lims, lims, '--', color='red', linewidth=1.5, label='y = x')

ax.set_xlim(lims)
ax.set_ylim(lims)

ax.set_xlabel("Mean Ratio")
ax.set_ylabel("Standard Deviation")
ax.set_title("NKX2-2 TREBL Experiment: Bio Reps Aggregated\n(Raw Barcodes\n& Directional Clustered UMIs)")

# ax.set_xscale('log')
# ax.set_yscale('log')
```

# After whitelisting barcodes

```python
whitelist_results_with_step1 = pd.read_csv("../../output/TL4_umi_results_whitelist_concat/results_with_step1.csv", index_col = 0)
whitelist_results_with_step1 
```

```python
whitelist_results_with_step1["complex_ratio"] = whitelist_results_with_step1["RTBC_umi_count_complex"] / whitelist_results_with_step1["AD_umi_count_complex"]
whitelist_results_with_step1
```

```python
whitelist_results_with_step1_agg = whitelist_results_with_step1[["AD", "complex_ratio", "name_RPTR"]].groupby(["AD",  "name_RPTR"]).agg([np.mean, np.std]).reset_index()
whitelist_results_with_step1_agg.columns = ["AD", "name_RPTR", "mean", "std"]
whitelist_results_with_step1_agg
```

```python
# Most ADs have reasonable error
import seaborn as sns
sns.set_context('talk')
plt.figure(dpi = 300)
ax = sns.scatterplot(data = whitelist_results_with_step1_agg, 
                     x = "mean", 
                     y = "std", 
                     edgecolor = 'none', alpha = 0.5)
sns.despine()

# Add y = x line
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
lims = [min(xmin, ymin), max(xmax, ymax)]
ax.plot(lims, lims, '--', color='red', linewidth=1.5, label='y = x')

ax.set_xlim(lims)
ax.set_ylim(lims)

ax.set_xlabel("Mean Ratio")
ax.set_ylabel("Standard Deviation")
ax.set_title("NKX2-2 TREBL Experiment: Bio Reps Aggregated\n(Density Whitelisted Barcodes\n& Directional Clustered UMIs)")

# ax.set_xscale('log')
# ax.set_yscale('log')
```

```python
whitelist_results_with_step1_agg[whitelist_results_with_step1_agg["std"] > whitelist_results_with_step1_agg["mean"]].sort_values(by = 'std')
```

```python
high_std_AD = whitelist_results_with_step1_agg[whitelist_results_with_step1_agg["std"] > whitelist_results_with_step1_agg["mean"]].sort_values(by = 'std')["AD"].iloc[3]
high_std_AD
```
