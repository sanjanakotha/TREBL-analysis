```python
import pickle

with open("../../data/ADBCs_correc_dict.pkl", "rb") as f:
    ADBCs_correc_dict = pickle.load(f)

with open("../../data/RPTR_correc_dict.pkl", "rb") as f:
    RPTR_correc_dict = pickle.load(f)

with open("../../data/AD_correc_dict.pkl", "rb") as f:
    AD_correc_dict = pickle.load(f)
```

# AD UMI Count

```python
all_complex_AD_results_df_unclustered = pd.read_csv(os.path.join(output_path, f"all_complex_AD_results.csv"), index_col = 0)
all_complex_AD_results_df_unclustered
```

```python
all_complex_AD_results_df_unclustered["number"].value_counts()
```

```python
all_complex_AD_results_df = all_complex_AD_results_df_unclustered.copy()
all_complex_AD_results_df["ADBC_concat"] = all_complex_AD_results_df["ADBC_concat"].map(lambda x: ADBCs_correc_dict.get(x, x))
all_complex_AD_results_df = (
    all_complex_AD_results_df
    .groupby(["ADBC_concat", "number"], as_index=False)
    .agg({
        "AD_umi_count": "sum",
        **{col: "first" for col in all_complex_AD_results_df.columns if col not in ["ADBC_concat", "AD_umi_count"]}
    })
)
all_complex_AD_results_df
```

```python
all_complex_AD_results_df["number"].value_counts()
```

# RPTR UMI Count

```python
all_complex_RPTR_results_df_unclustered = pd.read_csv(os.path.join(output_path, f"all_complex_RPTR_results.csv"), index_col = 0)
all_complex_RPTR_results_df_unclustered
```

```python
all_complex_RPTR_results_df_unclustered["number"].value_counts()
```

```python
all_complex_RPTR_results_df = all_complex_RPTR_results_df_unclustered.copy()
all_complex_RPTR_results_df["RTBC"] = all_complex_RPTR_results_df["RTBC"].map(lambda x: RPTR_correc_dict.get(x, x))
all_complex_RPTR_results_df = (
    all_complex_RPTR_results_df
    .groupby(["RTBC", "number"], as_index=False)
    .agg({
        "RTBC_umi_count": "sum",
        **{col: "first" for col in all_complex_RPTR_results_df_unclustered.columns if col not in ["RTBC", "RTBC_umi_count"]}
    })
)
all_complex_RPTR_results_df
```

```python
all_complex_RPTR_results_df["number"].value_counts()
```

```python
# Add complex AD UMI counts
step1_map_with_AD = pd.merge(step1_map, all_complex_AD_results_df, on = "ADBC_concat").drop_duplicates()
step1_map_with_AD
```

```python
# Add complex RPTR UMI counts
step1_map_with_AD_RPTR = pd.merge(step1_map_with_AD, all_complex_RPTR_results_df, on = ["RTBC", "number"], suffixes = ("_AD", "_RPTR"))
step1_map_with_AD_RPTR
```

```python
step1_map_with_AD_RPTR_std_devs = step1_map_with_AD_RPTR[["AD", "number", "ratio"]].groupby(["AD", "number"]).agg(np.std).sort_values(by = 'ratio').dropna().reset_index()
step1_map_with_AD_RPTR_std_devs  = step1_map_with_AD_RPTR_std_devs.rename(columns = {"ratio" : "ratio_stdev"})
step1_map_with_AD_RPTR_std_devs
```
