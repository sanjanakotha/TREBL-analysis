```python
import pandas as pd
from Bio.Seq import Seq
import numpy as np
```

```python
whitelist_results_with_step1 = pd.read_csv("../../output/TL4_umi_results_whitelist/results_with_step1.csv")
whitelist_results_with_step1["AD_prot"] = whitelist_results_with_step1["AD"].map(lambda s: str(Seq(s).translate()))
whitelist_results_with_step1
```

```python
whitelist_results_with_step1["complex_ratio"] = whitelist_results_with_step1["RTBC_umi_count_complex"] / whitelist_results_with_step1["AD_umi_count_complex"]
whitelist_results_with_step1
```

```python
whitelist_results_with_step1_agg = whitelist_results_with_step1[["AD", "complex_ratio", "name_RPTR"]].groupby(["AD",  "name_RPTR"]).agg([np.mean, np.std]).reset_index()
whitelist_results_with_step1_agg.columns = ["AD", "name_RPTR", "mean", "std"]
whitelist_results_with_step1_agg
```

```python
# Most ADs have reasonable error
import seaborn as sns
ax = sns.scatterplot(data = whitelist_results_with_step1_agg, x = "mean", y = "std", edgecolor = 'none')
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
whitelist_results_with_step1_agg[whitelist_results_with_step1_agg["std"] > whitelist_results_with_step1_agg["mean"]].sort_values(by = 'std')
```

```python
high_std_AD = whitelist_results_with_step1_agg[whitelist_results_with_step1_agg["std"] > whitelist_results_with_step1_agg["mean"]].sort_values(by = 'std')["AD"].iloc[8]
high_std_AD
```

```python
whitelist_results_with_step1_4 = whitelist_results_with_step1[whitelist_results_with_step1["name_RPTR"] == "r4_RNA_RTBC_S28"]
whitelist_results_with_step1_4[whitelist_results_with_step1_4["AD"] == high_std_AD]
```
