```python
def get_counts_per_concat(grouped, col_names, should_print = True):
    # Need quality BCs since all must be the same length
    grouped_qual = grouped
    for col_name in col_names:
        grouped_qual = grouped_qual[grouped_qual[col_name + "_qual"]]

    # Concatenate columns
    concat_col_name = "_".join(col_names) + "_concat"
    grouped_qual[concat_col_name] = ""
    for col_name in col_names:
        grouped_qual[concat_col_name] = grouped_qual[concat_col_name] + grouped_qual[col_name]

    # Get counts per group
    counts_per_concat = (
        grouped_qual[[concat_col_name, "count", "AD", "Designed"]]
        .groupby(concat_col_name, as_index=True)
        .agg({
            "count": "sum",        # sum counts
            "AD": "first",         # keep first AD (they should all be the same)
            "Designed": "first",     # keep True if any True
        })
        .sort_values(by = "count", ascending = True)
    )    
    
    if should_print:
        print(f"Starting with {len(counts_per_concat)} unique {concat_col_name}s")

    return counts_per_concat
    
def return_clustered(grouped, col_names, clustering_type):
    """
    Cluster concatenated barcode sequences using UMI-based methods.

    Parameters
    ----------
    grouped : pandas.DataFrame
        DataFrame containing barcode columns, their quality flags (<col>_qual),
        and a "count" column.
    col_names : list of str
        Names of barcode columns to concatenate and cluster.
    clustering_type : {'directional', 'adjacency'}
        Type of clustering algorithm to use.

    Returns
    -------
    pandas.DataFrame
        Clustered barcode groups with the concatenated barcode column and
        a 'group_size' column indicating cluster size.

    """
    concat_col_name = "_".join(col_names) + "_concat"
    
    counts_per_concat = get_counts_per_concat(grouped, col_names)
    display(counts_per_concat)
    
    # Convert counts to dict
    counts_per_concat_dict = counts_per_concat[["count"]].to_dict()['count']
    
    # Convert all keys to bytes safely (if not already bytes)
    counts_per_concat_dict_bytes = {
        (k.encode('utf-8') if isinstance(k, str) else k): v
        for k, v in counts_per_concat_dict.items()
    }

    if clustering_type == 'directional':
        print("Starting directional clustering...")
        directional_clusterer = UMIClusterer(cluster_method="directional")
        clustered = directional_clusterer(counts_per_concat_dict_bytes, threshold=1)
    else:
        print("Starting adjacency clustering...")
        adjacency_clusterer = UMIClusterer(cluster_method="adjacency")
        clustered = adjacency_clusterer(counts_per_concat_dict_bytes, threshold=1)

    clustered_df = pd.DataFrame(clustered)
    display(clustered_df)
    clustered_df['group_size'] = clustered_df.notna().sum(axis=1)
    # Convert all bytes to strings safely
    clustered_df = clustered_df.applymap(
        lambda x: x.decode() if isinstance(x, bytes) else x
    )
    clustered_df = clustered_df.rename(columns = {0: concat_col_name})
        
    return clustered_df

    #return counts_per_concat_dict_bytes
```

```python
def get_correc_dict(col_name, clustered_df):
    # Keep only the numbered columns (all columns except the concat name + group_size)
    numbered_cols = [col for col in clustered_df.columns if col not in [col_name, "group_size"]]
    
    # Melt to long format
    cluster_mapping = (
        clustered_df.melt(
            id_vars=[col_name], 
            value_vars=numbered_cols, 
            value_name="barcode"
        )
        .dropna(subset=["barcode"])  # remove Nones
        .drop(columns=["variable"])
    )
    cluster_mapping = cluster_mapping.rename(columns = {col_name : "corrected", "barcode" : "original"})
    return dict(zip(cluster_mapping["original"], cluster_mapping["corrected"]))
```

```python
def row_i_counts(grouped, col_names, clustered_df, i):
    # Get the row as a list, dropping NaNs
    row_seqs = clustered_df.loc[i].dropna().tolist()

    concat_col_name = "_".join(col_names) + "_concat"
    
    counts_per_concat = get_counts_per_concat(grouped, col_names, should_print = False).reset_index()
    
    # Filter counts_per_AD where 'AD' is in row_ADs
    return counts_per_concat[counts_per_concat[concat_col_name].isin(row_seqs)].sort_values(by = 'count', ascending = False)
    
```

```python
import itertools

def hamming_distance(s1, s2):
    s1, s2 = str(s1), str(s2)  # ensure both are strings
    # pad shorter string if necessary
    # if len(s1) != len(s2):
    #     max_len = max(len(s1), len(s2))
    #     s1 = s1.ljust(max_len, "-")
    #     s2 = s2.ljust(max_len, "-")
    return sum(ch1 != ch2 for ch1, ch2 in zip(s1, s2))

def display_row_i_counts_concat_ADBC(i):
    # Get the row as a list, dropping NaNs
    row_ADBCs = adjacency_clustered_AD_barcodes_df.iloc[i].dropna().tolist()
    
    # Filter relevant rows
    relev_rows = counts_per_concat_ADBC[
        counts_per_concat_ADBC['concat_ADBC'].isin(row_ADBCs)
    ].sort_values(by='count', ascending=False).drop_duplicates()

    display(relev_rows)

    # Extract unique ADs and coerce to string
    ADs = relev_rows["AD"].dropna().drop_duplicates().astype(str).tolist()

    if len(ADs) > 1:
        top_AD = ADs[0]
        print(f"Multiple ADs! Using top AD for comparison.")
        dist_data = []
        for ad in ADs[1:]:
            dist = hamming_distance(top_AD, ad)
            dist_data.append((top_AD, ad, dist))
        dist_df = pd.DataFrame(dist_data, columns=["Top_AD", "Other_AD", "Hamming_Distance"])
        display(dist_df)
    else:
        print("Only one unique AD — no distances to show.")

# Example usage: check a few random high-group-size rows
for i in adjacency_clustered_AD_barcodes_df.sort_values(by='group_size', ascending=False).head(1000).sample(5).index:    
    display_row_i_counts_concat_ADBC(i)
    print("----")
```
