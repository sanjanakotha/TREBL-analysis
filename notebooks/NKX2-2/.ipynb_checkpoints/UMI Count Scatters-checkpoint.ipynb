```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import duckdb
from concurrent.futures import ThreadPoolExecutor, as_completed
import dask.dataframe as dd
import sys

sys.path.append("../../scripts")
%aimport initial_map
%aimport map_refiner
%aimport complexity
%aimport finder
%aimport preprocess
%aimport complexity
    
import glob
from tqdm import tqdm  # progress bar
```

```python
simple_AD_results = pd.read_csv("../../output/TL4_umi_results/all_simple_AD_results.csv")
complex_AD_results = pd.read_csv("../../output/TL4_umi_results/all_complex_AD_results.csv")
merged_AD_results = pd.merge(simple_AD_results, complex_AD_results, on = ["name", "number", "ADBC_concat"], suffixes = ("_simple", "_complex"))
merged_AD_results
```

```python
merged_AD_results["duplicates_removed"] = merged_AD_results["AD_umi_count_simple"] - merged_AD_results["AD_umi_count_complex"]
merged_AD_results
```

```python
simple_RPTR_results = pd.read_csv("../../output/TL4_umi_results/all_simple_RPTR_results.csv")
complex_RPTR_results = pd.read_csv("../../output/TL4_umi_results/all_complex_RPTR_results.csv")
merged_RPTR_results = pd.merge(simple_RPTR_results, complex_RPTR_results, on = ["name", "number", "RTBC"], suffixes = ("_simple", "_complex"))
merged_RPTR_results
```

```python
merged_RPTR_results["duplicates_removed"] = merged_RPTR_results["RTBC_umi_count_simple"] - merged_RPTR_results["RTBC_umi_count_complex"]
merged_RPTR_results
```

```python
sns.set_context('talk')
fig, axes = plt.subplots(1, 2, figsize=(10,5), sharey=True, sharex = True)

# --- Plot 1: AD Results ---
sns.scatterplot(
    data=merged_AD_results,
    x="AD_umi_count_simple",
    y="duplicates_removed",
    edgecolor='none',
    alpha=0.25,
    ax=axes[0],
    color = sns.color_palette("tab10")[0]
)
sns.despine(ax=axes[0])
axes[0].set_xlabel("")
axes[0].set_ylabel("")
axes[0].set_title("AD BC")

# --- Plot 2: RPTR Results ---
sns.scatterplot(
    data=merged_RPTR_results,
    x="RTBC_umi_count_simple",
    y="duplicates_removed",
    edgecolor='none',
    alpha=0.25,
    ax=axes[1],
    color = sns.color_palette("tab10")[1]
)
sns.despine(ax=axes[1])
axes[1].set_xlabel("")
axes[1].set_ylabel("")
axes[1].set_title("Reporter BC")

fig.supxlabel("Simple UMI Count", y = 0.07, x = 0.55)
fig.supylabel("UMI Duplicates Removed", y = 0.55, x = 0.05)

plt.tight_layout()
```

```python
sns.set_context('talk')
fig, axes = plt.subplots(1, 2, figsize=(10,5), dpi = 300, sharex=True, sharey=True)

# --- Plot 1: AD Results ---
ax = axes[0]
sns.scatterplot(
    data=merged_AD_results,
    x="AD_umi_count_simple",
    y="AD_umi_count_complex",
    edgecolor='none',
    alpha=0.5,
    s=25,
    ax=ax
)
sns.despine(ax=ax)

lims = [
    min(ax.get_xlim()[0], ax.get_ylim()[0]),
    max(ax.get_xlim()[1], ax.get_ylim()[1])
]
ax.plot(lims, lims, '--', color='red', linewidth=1)
ax.set_xlim(lims)
ax.set_ylim(lims)
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_title("AD BC")

# --- Plot 2: RPTR Results ---
ax2 = axes[1]
sns.scatterplot(
    data=merged_RPTR_results,
    x="RTBC_umi_count_simple",
    y="RTBC_umi_count_complex",
    edgecolor='none',
    alpha=0.5,
    s=25,
    ax=ax2,
    color = sns.color_palette("tab10")[1]

)
sns.despine(ax=ax2)

lims2 = [
    min(ax2.get_xlim()[0], ax2.get_ylim()[0]),
    max(ax2.get_xlim()[1], ax2.get_ylim()[1])
]
ax2.plot(lims2, lims2, '--', color='red', linewidth=1)
ax2.set_xlim(lims2)
ax2.set_ylim(lims2)
ax2.set_xlabel("")
#ax2.set_ylabel("Directional RPTR UMI Count")
ax2.set_title("Reporter BC")

fig.supxlabel("Simple UMI Count", y = 0.07, x = 0.55)
fig.supylabel("Directional UMI Count", y = 0.55, x = 0.05)

plt.tight_layout()
```
