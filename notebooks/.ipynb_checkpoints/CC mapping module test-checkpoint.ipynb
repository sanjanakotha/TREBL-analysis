```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get loss table with '% of previous step'
df = refiner.save_loss_table()

# Set plot style
sns.set(style="white", context='talk')

# Create figure and axes
fig, ax1 = plt.subplots(figsize=(8, 4), dpi=300)

# Horizontal bar plot
sns.barplot(x="num_rows", y="description", data=df, ax=ax1, color="skyblue")

# Format y-axis labels (map descriptions)
ax1.set_yticklabels(
    df["map"]
    .str.split("_").str[1:]
    .str.join(" ")
    .str.split(" combinations").str[0]
    .str.replace(" ", "\n")
)

# Add labels at the end of each bar with % of previous step
max_rows = df["num_rows"].max()
for i, row in df.iterrows():
    x = row["num_rows"] + 0.02 * max_rows  # 2% offset to the right
    pct_prev = f"{row['% of previous step']}%" if row['% of previous step'] is not None else "—"
    ax1.text(x, i, f'{row["num_rows"]:,} ({pct_prev})', va='center', fontsize='small')

# Axis labels
ax1.set_xlabel("Num Rows")
ax1.set_ylabel("Map Step")

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

```python
import duckdb 
from itertools import combinations
import seaborn as sns
import matplotlib.pyplot as plt

con = duckdb.connect("/global/scratch/projects/fc_mvslab/OpenProjects/Sanjana/TREBL/duckdb/analysis_cc_test_3.db")
```

```python
# Plot TBBS count hist
```

```python
# Columns to consider
cols = [ "AD", "ADBC2", "HawkBCs", "RTBC"]

n = len(cols)
fig, axes = plt.subplots(n, n, figsize=(12, 12),
                         sharex='col', sharey='row')
sns.set(style="ticks")

for i, row_var in enumerate(cols):
    for j, col_var in enumerate(cols):
        ax = axes[i, j]
        if i > j:  # bottom-left half only
            query = f"""
            SELECT {col_var} AS x, COUNT(DISTINCT {row_var}) AS unique_count
            FROM map3
            GROUP BY {col_var}
            """
            df_pair = con.execute(query).fetchdf()
            sns.histplot(df_pair['unique_count'], bins=20, ax=ax, color='steelblue')

            # Only label outer plots
            if j == 0:
                ax.set_ylabel(f'Unique {row_var} count', fontsize=10)
            if i == n-1:
                ax.set_xlabel(f'{col_var}', fontsize=10)

            ax.set_xscale('log')
            ax.set_yscale('log')
        else:
            ax.axis('off')

sns.despine()
plt.tight_layout()
plt.show()
```
