```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
AD_BC_results = pd.read_csv("../../data/ChopTFs/Spike-in/ADBC_results.csv", index_col = 0)
AD_BC_results["time"] = AD_BC_results["name"].str.split("_").str[3].astype(int)
AD_BC_results  = AD_BC_results[["AD", "AD_BC", "time", "AD_umi_count_simple"]]
AD_BC_results
```

```python
RT_BC_results = pd.read_csv("../../data/ChopTFs/Spike-in/RPTR_BC_results.csv", index_col = 0)
RT_BC_results["time"] = RT_BC_results["name"].str.split("_").str[3].astype(int)
RT_BC_results  = RT_BC_results[["RPTR_BC", "time", "RPTR_BC_umi_count_simple"]]
RT_BC_results
```

```python
step1_map = pd.read_csv("../../data/ChopTFs/ChopTF_step1_20260114.csv")
step1_map = step1_map[["AD", "AD_BC", "RPTR_BC"]]
step1_map
```

```python
step1_map["AD"].value_counts()
```

```python
step1_map_with_AD = pd.merge(step1_map, AD_BC_results)
step1_map_with_AD
```

```python
step1_map_with_AD_RT = pd.merge(step1_map_with_AD, RT_BC_results)
step1_map_with_AD_RT
```

```python
step1_map_with_AD_RT["activity"] = step1_map_with_AD_RT["RPTR_BC_umi_count_simple"] / step1_map_with_AD_RT["AD_umi_count_simple"]
step1_map_with_AD_RT
```

```python
step1_map_with_AD_RT = step1_map_with_AD_RT.drop(columns = ["AD_umi_count_simple", "RPTR_BC_umi_count_simple"])
step1_map_with_AD_RT
```

```python
grouped = step1_map_with_AD_RT.groupby(["AD", "time"])["activity"].agg(["mean", "std"]).reset_index()
grouped
```

```python
pivoted_grouped_activity = grouped.pivot(index="AD", columns="time", values="mean")
pivoted_grouped_activity['na_count'] = pivoted_grouped_activity.isna().sum(axis=1)
pivoted_grouped_activity
```

```python
pivoted_grouped_activity.drop(columns="na_count").to_csv("../../output/ChopTFs_spike-in_activities.csv")
```

```python
pivoted_grouped_sd = grouped.pivot(index="AD", columns="time", values="std")
pivoted_grouped_sd['na_count'] = pivoted_grouped_sd.isna().sum(axis=1)
pivoted_grouped_sd
```

```python
pivoted_grouped_activity.sort_values(by = 'na_count')
```

```python
# Filter rows with no nulls (na_count == 0)
no_nulls = pivoted_grouped_activity[pivoted_grouped_activity['na_count'] == 0]

# Filter rows with no nulls (na_count == 0) and select the first 10 rows
no_nulls_subset = no_nulls.sample(10)

# Select columns corresponding to time points (0-60)
time_columns = [col for col in pivoted_grouped_activity.columns if isinstance(col, int)]

# Plot lineplot with 10 subplots
fig, axes = plt.subplots(2,5, figsize=(12,4), sharex=True, dpi = 300)
axes = axes.flatten()

for i, (index, row) in enumerate(no_nulls_subset.iterrows()):
    ax = axes[i]
    ax.plot(time_columns, row[time_columns], marker='o')
    ax.set_yscale('log')

fig.supylabel("Activity")
fig.supxlabel("Time")

plt.suptitle("Chop TFs Spike-in")
sns.despine()
plt.tight_layout()
plt.show()
```
